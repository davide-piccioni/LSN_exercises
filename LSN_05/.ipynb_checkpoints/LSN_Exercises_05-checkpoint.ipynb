### <font color="red">Exercise 05.1: generating probability density of hydrogen wavefunctions using Metropolis algorithm <font>

In this laboratory session we want to exploit the Metropolis algorithm to generate some probability density; we will use as an example the square modulus of the eigenfunctions of a spinless hydrogen atom.

This turns out to be a well defined probability density since in quantum physics, for a non-relativistic single particle without spin, its state is completely described by the wave function $\Psi(\vec{r},t)$, where $\vec{r}$ is position and $t$ is time, which is a complex-valued probability amplitude. Moreover the square modulus of the wave function, $|\Psi(\vec{r},t)|^2$, is interpreted as the probability density that the particle is at $\vec{r}$ at time $t$.

The case of the hydrogen atom consists of an electron interacting with a proton by coulombic interaction; it can be resolved exactly and the wave functions of its eigenstates are expressed in terms of spherical harmonics and generalized Laguerre polynomials. Using spherical coordinates, the wave functions can be written as:

$$
\Psi_{n,l,m}(r,\theta,\phi)=
\sqrt{\left(\frac{2}{na_0}\right)^3 \frac{(n-l-1)!}{2n[(n+l)!]}}
e^{-r/na_0}\left(\frac{2r}{na_0}\right)^l L_{n-l-1}^{2l+1}\left(\frac{2r}{na_0}\right)
Y_l^m(\theta,\phi)
$$

where $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529$ nm is the Bohr radius, $L_{n-l-1}^{2l+1}$ are the generalized Laguerre polynomials of degree $n-l-1$, $Y_l^m(\theta,\phi)$ are spherical harmonics, $n=1,2,...$ is the principal quantum number, $l=0,1, ..., n-1$ the azimuthal quantum number, $m=-l, -l+1, ..., +l-1, +l$  the magnetic quantum number.

We will be interested in the ground state wave function:

$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{1}{a_0^{3/2}\sqrt{\pi}} e^{-\frac{r}{a_0}} 
$$

We will also reproduce the probability density of one of the three $2p$ excited states:

$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{1}{8 a_0^{5/2}}\sqrt{\frac{2}{\pi}} r e^{-\frac{r}{2 a_0}} \cos(\theta)
$$

Expectation values for the radius turns out to be exactly:

$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
</span>

We specify that we are going to use Bohr radius units, $a_0$ for distances.

### Method

We are goin to use the Metropolis algorithm to sample $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$ in Cartesian coordinates, first using an uniform transition probability $T(\vec{x}|\vec{y})$ and then using a multivariate normal transition probability $T(\vec{x}|\vec{y})$. 

Then we'll take advantage of the sampled positions to estimate $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$. As always we use data blocking to calculate uncertainties.

### Ground state

We set the step of the <em> uniform transition probability </em> equal to $1.2 \ a_0$ in order to obtain an acceptance rate of around 50%. We take $\left( 1,1,1\right)$ as a starting point and calculate the following mean value for the radius, compatible with the expected value.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

M=1000000             # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N


x = np.arange(N)      # [0,1,2,...,N-1]

sum_prog, err_prog = np.loadtxt("./5.1/raggi_1s_unif.dat", usecols=(0,1), delimiter=' ', unpack='true')    
    
x*=L # Number of throws = block * (Number of throws in each block)
fig=plt.figure(figsize=(14, 8))
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.title('Mean value for the radius, hydrogen ground state, uniform transition', fontsize=18)
plt.xlabel('#throws', fontsize=14)
plt.ylabel('<r> a', fontsize=14)
plt.axhline(y=1.5, color='r', linestyle='-')
plt.grid(True)
plt.show()
```

If we switch the <em> transition probability </em> $T(\vec{x}|\vec{y})$ to a <em> gaussian </em> one, the result is still compatible with the expected valute setting the initial position in the same place and choosing $0.75 \ a_0$ as the standard deviation of the gaussian centered at each point.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

M=1000000             # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N


x = np.arange(N)      # [0,1,2,...,N-1]

sum_prog, err_prog = np.loadtxt("./5.1/raggi_1s_gauss.dat", usecols=(0,1), delimiter=' ', unpack='true')    
    
x*=L # Number of throws = block * (Number of throws in each block)
fig=plt.figure(figsize=(14, 8))
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.title('Mean value for the radius, hydrogen ground state, gaussian transition', fontsize=18)
plt.xlabel('#throws', fontsize=14)
plt.ylabel('<r> a', fontsize=14)
plt.axhline(y=1.5, color='r', linestyle='-')
plt.grid(True)
plt.show()
```

We then represent the sampled points distributed into the 3D space:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

M=100010

x, y, z = np.loadtxt("./5.1/risultati_1s.dat", usecols=(0,1,2), delimiter=' ', unpack='true')    

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z, c=z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

### Observations

We observe that if we start from very far from the origin, like in $\left( 100,1,1\right)$ it takes quite a lot of steps to reach the equilibrium; in this case it takes around $50$ steps.

The number of blocks we have chosen implies we have $10^4$ Monte Carlo Metropolis steps in each block. Since the position can be considered quite uncorrelated (steps of $1.2 \ a_0$ where the mean value of $\left\langle r \right\rangle$ is $1.5 \ a_0$ means we can move quite a lot through the space with a single move), changing the number of blocks shouldn't change much the results. For example, here is $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ with uniform transition probability using $10000$ blocks instead of $100$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

M=1000000             # Total number of throws
N=10000               # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N


x = np.arange(N)      # [0,1,2,...,N-1]

sum_prog, err_prog = np.loadtxt("./5.1/raggi_1s_unif_1000blocks.dat", usecols=(0,1), delimiter=' ', unpack='true')    
    
x*=L # Number of throws = block * (Number of throws in each block)
fig=plt.figure(figsize=(14, 8))
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.title('Mean value for the radius, hydrogen ground state, uniform transition', fontsize=18)
plt.xlabel('#throws', fontsize=14)
plt.ylabel('<r> a', fontsize=14)
plt.axhline(y=1.5, color='r', linestyle='-')
plt.grid(True)
plt.show()
```

### Excited state

In this case we try to generate a probability distribution related to the state $\Psi_{2,1,0}$. We set the step of the <em> uniform transition probability </em> equal to $2.8 \ a_0$ in order to obtain an acceptance rate of around 50%. We take $\left( 1,1,1\right)$ as a starting point and calculate the following mean value for the radius, compatible with the expected value.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

M=1000000             # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N


x = np.arange(N)      # [0,1,2,...,N-1]

sum_prog, err_prog = np.loadtxt("./5.1/raggi_2p_unif.dat", usecols=(0,1), delimiter=' ', unpack='true')    
    
x*=L # Number of throws = block * (Number of throws in each block)
fig=plt.figure(figsize=(14, 8))
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.title('Mean value for the radius, hydrogen 2p excited state, uniform transition', fontsize=18)
plt.xlabel('#throws', fontsize=14)
plt.ylabel('<r>', fontsize=14)
plt.axhline(y=5, color='r', linestyle='-')
plt.grid(True)
plt.show()
```

Then, one more time we switch the <em> transition probability </em> $T(\vec{x}|\vec{y})$ to a <em> gaussian </em> one and, setting the initial position in the same place, choose $1.8 \ a_0$ as the standard deviation of the gaussian centered at each point in order to get an acceptance rate around 50%. The result is still compatible with the expected value.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

M=1000000             # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N


x = np.arange(N)      # [0,1,2,...,N-1]

sum_prog, err_prog = np.loadtxt("./5.1/raggi_2p_gauss.dat", usecols=(0,1), delimiter=' ', unpack='true')    
    
x*=L # Number of throws = block * (Number of throws in each block)
fig=plt.figure(figsize=(14, 8))
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.title('Mean value for the radius, hydrogen 2p excited state, gaussian transition', fontsize=18)
plt.xlabel('#throws', fontsize=14)
plt.ylabel('<r>', fontsize=14)
plt.axhline(y=5, color='r', linestyle='-')
plt.grid(True)
plt.show()
```
