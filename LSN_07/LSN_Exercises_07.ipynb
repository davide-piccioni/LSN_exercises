### <font color="red">Exercise 07.1: Studying the autocorrelation functions <font>

In this notebook we will use a Molecular Dynamics code that performs simulations by moving the particles using the Metropolis algorithm; the molecules are subjected to Lennard-Jones potential.

For this first exercise we will study the autocorrelation functions and the changing of statistical uncertainties due to different length of the blocks in which our data are divided.

First we perform Monte Carlo simulations and produce as output the instantaneous values of potential energy per particle, $U/N$, and pressure, $P$, being sure that the system has reached equilibrium. These data has to be produced for the following systems:
- solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
- liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
- gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

Then we process the data inside the notebook in order to obtain the autocorrelation functions of $U/N$ and $P$, these functions are then fitted with a decreasing exponential giving the Monte Carlo correlation time as a result.

At last we observe the final statistical uncertainties of the expectation value of $U/N$ and $P$ while dividing the raw data in blocks of different sizes and using as uncertainty the standard deviation of the mean. We will represent the uncertainty as a function of the length of the blocks, from $L=10$ to $L=5\times 10^3$.

## Solid

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def f(x,a,b):  # defining the fitting function
    return a*np.exp(-x/b)

raw_potential= np.loadtxt("./MonteCarlo_NVT/Inst_potential_solid.dat", usecols=(0), delimiter=' ', unpack='true')   

N=raw_potential.size
k=int(N/100)

x=np.arange(k)
chi=np.zeros(k)

variance=0
sum_simple=0
sum_quad=0
for i in range(N):
    sum_simple+=raw_potential[i]
    sum_quad+=raw_potential[i]**2
variance=sum_quad/N-(sum_simple/N)**2

print('1 of 2')
for i in range(k):
    if((i*10)%k==0):
        print(int((i*100)/k),end='%, ')
    sum_1=0
    sum_2=0
    sum_3=0
    for j in range(N-i):
        sum_1+=raw_potential[j]*raw_potential[j+i]
        sum_2+=raw_potential[j]
        sum_3+=raw_potential[j+i]
    sum_1=sum_1/float(N-i)
    sum_2=sum_2/float(N-i)
    sum_3=sum_3/float(N-i)
    chi[i]=(sum_1-sum_2*sum_3)/variance
print("100%. \n")
    
p_opt, p_cov = curve_fit(f, x, chi)
y_fit = f(x,p_opt[0],p_opt[1])
    
fig=plt.figure(figsize=(14, 8))
plt.subplot(1, 2, 1)
plt.title('Autocorrelation Potential', fontsize=18)
plt.plot(x,chi)
plt.plot(x,y_fit) # plotting fitted function
plt.xlabel('step', fontsize=16)
plt.ylabel('χ', fontsize=16)
plt.grid(True)

tau_pot=p_opt[1];

raw_pressure= np.loadtxt("./MonteCarlo_NVT/Inst_pre_solid.dat", usecols=(0), delimiter=' ', unpack='true')   

N=raw_pressure.size
k=int(N/100)

x=np.arange(k)
chi=np.zeros(k)

variance=0
sum_simple=0
sum_quad=0
for i in range(N):
    sum_simple+=raw_pressure[i]
    sum_quad+=raw_pressure[i]**2
variance=sum_quad/N-(sum_simple/N)**2

print('2 of 2')
for i in range(k):
    if((i*10)%k==0):
        print(int((i*100)/k),end='%, ')
    sum_1=0
    sum_2=0
    sum_3=0
    for j in range(N-i):
        sum_1+=raw_pressure[j]*raw_pressure[j+i]
        sum_2+=raw_pressure[j]
        sum_3+=raw_pressure[j+i]
    sum_1=sum_1/float(N-i)
    sum_2=sum_2/float(N-i)
    sum_3=sum_3/float(N-i)
    chi[i]=(sum_1-sum_2*sum_3)/variance
print('100%.')

p_opt, p_cov = curve_fit(f, x, chi)
y_fit = f(x,p_opt[0],p_opt[1])

plt.subplot(1, 2, 2)
plt.title('Autocorrelation Pressure', fontsize=18)
plt.plot(x,chi)
plt.plot(x,y_fit) # plotting fitted function
plt.xlabel('step', fontsize=16)
plt.ylabel('χ', fontsize=16)
plt.grid(True)
plt.show()

tau_pres=p_opt[1];

print("Tau_potential =", end=" ");
print(tau_pot)

print("Tau_pressure =", end=" ");
print(tau_pres)
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def error(AV,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n)

#raw_potential= np.loadtxt("./MonteCarlo_NVT/Inst_potential.dat", usecols=(0), delimiter=' ', unpack='true')   

vet_incert=np.zeros(100)

for L in range(10,5010,50): # Number of throws in each block, please use for M a multiple of N
    if (L-10)%1000==0:
        print(int((L-10)/1000),end=" of ")
        print(5,end=", ")
    M=raw_potential.size                 
    N=int(M/L)            # Number of blocks

    x = np.arange(N)      # [0,1,2,...,N-1]
    ave = np.zeros(N)
    av2 = np.zeros(N)
    sum_prog = np.zeros(N)
    su2_prog = np.zeros(N)
    err_prog = np.zeros(N)

    for i in range(N):
        sum = 0
        for j in range(L):
            k = j+i*L
            sum += raw_potential[k]
        ave[i] = sum/L       # r_i 
        av2[i] = (ave[i])**2 # (r_i)^2 

    for i in range(N):
        for j in range(i+1):
            sum_prog[i] += ave[j] # SUM_{j=0,i} r_j
            su2_prog[i] += av2[j] # SUM_{j=0,i} (r_j)^2
        sum_prog[i]/=(i+1) # Cumulative average
        su2_prog[i]/=(i+1) # Cumulative square average
        err_prog[i] = error(sum_prog,su2_prog,i) # Statistical uncertainty
    
    vet_incert[int((L-10)/50)]=err_prog[-1]

print("5 of 5.")

x=np.arange(10,5010,50) # Number of throws = block * (Number of throws in each block)
fig=plt.figure(figsize=(14, 8))
plt.title('Potential: error as a function of the number of elements per block, L', fontsize=18)
plt.plot(x,vet_incert)
plt.xlabel('L')
plt.ylabel('Uncertainty')
plt.grid(True)

vet_incert=np.zeros(20)

for L in range(10,110,5): # Number of throws in each block, please use for M a multiple of N
    if (L-10)%20==0:
        print(int((L-10)/20),end=" of ")
        print(5,end=", ")
    M=raw_potential.size                 
    N=int(M/L)            # Number of blocks

    x = np.arange(N)      # [0,1,2,...,N-1]
    ave = np.zeros(N)
    av2 = np.zeros(N)
    sum_prog = np.zeros(N)
    su2_prog = np.zeros(N)
    err_prog = np.zeros(N)

    for i in range(N):
        sum = 0
        for j in range(L):
            k = j+i*L
            sum += raw_potential[k]
        ave[i] = sum/L       # r_i 
        av2[i] = (ave[i])**2 # (r_i)^2 

    for i in range(N):
        for j in range(i+1):
            sum_prog[i] += ave[j] # SUM_{j=0,i} r_j
            su2_prog[i] += av2[j] # SUM_{j=0,i} (r_j)^2
        sum_prog[i]/=(i+1) # Cumulative average
        su2_prog[i]/=(i+1) # Cumulative square average
        err_prog[i] = error(sum_prog,su2_prog,i) # Statistical uncertainty
    
    vet_incert[int((L-10)/5)]=err_prog[-1]
print("5 of 5.")

x=np.arange(10,110,5) # Number of throws = block * (Number of throws in each block)
fig=plt.figure(figsize=(14, 8))
plt.title('Potential: error as a function of the number of elements per block, L, detail L = [10,110]', fontsize=18)
plt.plot(x,vet_incert)
plt.xlabel('L')
plt.ylabel('Uncertainty')
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

L=1000            # Number of throws in each block, please use for M a multiple of N
N=100                  # Number of blocks
M=L*N                # Total number of throws

x = np.arange(N)      # [0,1,2,...,N-1]

epot_mean, epot_err = np.loadtxt("./MonteCarlo_NVT/output.pot_gas.0", usecols=(2,3), delimiter='   ', unpack='true')    
press_mean, press_err = np.loadtxt("./MonteCarlo_NVT/output.pre_gas.0", usecols=(2,3), delimiter='   ', unpack='true')       
iblt, r, av_gr, err_gr = np.loadtxt("./MonteCarlo_NVT/output.g_r_gas.0", usecols=(0,1,3,4), delimiter='   ', unpack='true')    

x*=L # Number of throws = block * (Number of throws in each block)

#E_pot
fig=plt.figure(figsize=(14, 8))
fig.subplots_adjust(hspace=0.4, wspace=0.2)
#Argon
plt.subplot(2, 1, 1)
plt.errorbar(x,epot_mean*120*1.38064853*(10**(-23)),yerr=epot_err*120*1.38064853*(10**(-23)), color='b')
plt.title('Potential Energy per particle of the system - Argon', fontsize=18)
plt.xlabel('#steps', fontsize=14)
plt.ylabel('U (J)', fontsize=14)
plt.grid(True)
#Krypton
plt.subplot(2, 1, 2)
plt.errorbar(x,epot_mean*164*1.38064853*(10**(-23)),yerr=epot_err*164*1.38064853*(10**(-23)), color='b')
plt.title('Potential Energy per particle of the system - Krypton', fontsize=18)
plt.xlabel('#steps', fontsize=14)
plt.ylabel('U (J)', fontsize=14)
plt.grid(True)
plt.show()
print('------------------------------------------------------------------------------------------------------------------')

#Pressure
fig=plt.figure(figsize=(14, 8))
fig.subplots_adjust(hspace=0.4, wspace=0.2)
#Argon
plt.subplot(2, 1, 1)
plt.errorbar(x,press_mean*120*1.38064853*(10**(-23))/((0.34*(10**(-9)))**3),yerr=press_err*120*1.38064853*(10**(-23))/((0.34*(10**(-9)))**3))
plt.title('Mean pressure of the system - Argon', fontsize=18)
plt.xlabel('#steps', fontsize=14)
plt.ylabel('P (Pa)', fontsize=14)
plt.grid(True)
#Krypton
plt.subplot(2, 1, 2)
plt.errorbar(x,press_mean*164*1.38064853*(10**(-23))/((0.364*(10**(-9)))**3),yerr=press_err*164*1.38064853*(10**(-23))/((0.364*(10**(-9)))**3))
plt.title('Mean pressure of the system - Krypton', fontsize=18)
plt.xlabel('#steps', fontsize=14)
plt.ylabel('P (Pa)', fontsize=14)
plt.grid(True)
plt.show()
print('------------------------------------------------------------------------------------------------------------------')

#g(r)
r_new=np.zeros(0);
av_new=np.zeros(0);
err_new=np.zeros(0);

for i in range(iblt.size):
    if iblt[i]==iblt[-1]:
        r_new=np.append(r_new,r[i]);
        av_new=np.append(av_new,av_gr[i]);
        err_new=np.append(err_new,err_gr[i]);

r, av_gr, err_gr = np.loadtxt("./MolecularDynamics_NVE/Gas/ave_gofr.dat", usecols=(0,1,2), delimiter=' ', unpack='true')    
        
fig=plt.figure(figsize=(14, 8))
plt.errorbar(r_new,av_new,yerr=err_new,label="Metropolis algoritm")
plt.errorbar(r,av_gr,yerr=err_gr,label="Verlet algoritm")
plt.title('Radial distribution function', fontsize=18)
plt.xlabel('r', fontsize=14)
plt.ylabel('g(r)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

### Conclusions

We can note that the radial distribution functions are in very good accord in the solid case; the accord becoming worse for the liquid and again for the gas. However these results indicate that both the algorithms are valid.

All the $g(r)$ being $0$ until they reach a first peak indicates that the Lennard-Jones potential makes it impossible to go below a certain distance between molecules.

From $g(r)$ we can distinguish the solid, with numerous peaks indicating an ordinated structure with a great number of neighbours per molecules (in the graph there are four peaks, from the second to the fifth neighbour), the liquid, with just two peaks at larger distance than the solid (indicating molecules bounded weakly) and the gas, with just one peak indicating that there are no bindings.
