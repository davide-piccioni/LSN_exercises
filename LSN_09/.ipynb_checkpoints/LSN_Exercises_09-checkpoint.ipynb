### <font color="red">Exercise 09.1: Genetic algorithms and the travelling salesman problem<font>

We had to design a genetic algorithm that could find a solution to the travelling salesman problem, altough there is often no certainty to have found the actual solution for this particular problem.

### The problem

The travelling salesman problem consists of findind the shortest close path that passes for $N$ given points, never passing twice over a point. So the idea is finding the particular permutation of the cities that describes the path minimizing the following function:
$$
L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|
$$

(with $\vec{x}_{N+1}=\vec{x}_1$).

### Strategy for resolution

We wrote a C++ code to try and solve the problem with a genetic algorithm. The idea behind it is the following:
- given $N$ cities with their positions in the plane, a population of $N^2$ random paths is invented;
- at each step the population is ordered according to the length of the paths;
- then, taking the members of this generation as parents, a new population is generated, the trick to simulate genetic evolution is to select the parents introducing a bias that favours paths shorter than others (the new generation is also subject to some mutation);

These steps are then performed in the same way generations after generations and, trusting in the power of the genetic algorithms, hopefully a good candidate appears spontaneusly in the population.

### Request

We were asked to test the code over 30 cities randomly placed **on** a circumference and 30 cities randomly placed **inside** a square. In the first case we can easily check the result, since we expect each city to be connected with its neighbours on the circumference in the shortest path case. In the second one we have no guarantees.

We showed the results via:

- a picture of $L$ of the best path and of $\langle L\rangle$, averaged on the best half of the population, as a function of the generations of the algorithm;
- a picture of the best path.

### Prevision and results

Obviously we expect the $L_{best}$ curve to be always below the $L_{average}$ and that's what we have found. 

We observe that the result for the best path in the circumference case is coherent with our prevision of finding neighbours points connected.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x,y = np.loadtxt("./9.1/Circonf/final_cities.dat", usecols=(0,1), unpack='true')    

x=np.append(x, x[0])
y=np.append(y, y[0])

fig=plt.figure(figsize=(14, 8))
plt.title('Circumference - Best path', fontsize=18)
plt.xlim((-1,1))
plt.ylim((-1,1))
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N=50                  # Number of generations

x = np.arange(50)      
x+=1                  # [1,2,...,N-1,N]


L_best = np.loadtxt("./9.1/Circonf/L_best.dat", usecols=(0), delimiter=' ', unpack='true')    
L_ave = np.loadtxt("./9.1/Circonf/L_average.dat", usecols=(0), delimiter=' ', unpack='true')    

fig=plt.figure(figsize=(14, 8))
plt.plot(x, L_best,color='r', label='L best')
plt.plot(x, L_ave,color='g', label='L average best half')
plt.legend()
plt.title('L at different generations - 30 cities on circumference', fontsize=18)
plt.xlabel('generation', fontsize=14)
plt.ylabel('L', fontsize=14)
plt.grid(True)
plt.show()
```

### Other cases

We tried to apply the algorithm also to 50 and 100 cities inside a square, these are the best elements of the populations. Note that it took more than $700$ generations for getting out from a local minimum in the 100 cities case.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x,y = np.loadtxt("./9.1/Square_50/final_cities.dat", usecols=(0,1), unpack='true')    

x=np.append(x, x[0])
y=np.append(y, y[0])

fig=plt.figure(figsize=(14, 8))
plt.title('50 Points inside a square - Best after 500 generations', fontsize=18)
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x,y = np.loadtxt("./9.1/Square_100/final_cities_700_gen.dat", usecols=(0,1), unpack='true')    

x=np.append(x, x[0])
y=np.append(y, y[0])

fig, ax1 = plt.subplots(1, 1, figsize=(14, 8))
plt.title('100 Points inside a square - Best after 700 generations', fontsize=18)
circle2 = plt.Circle((-0.3, 0.7), 0.2, color='r', fill=False)
ax1.add_artist(circle2)
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```
